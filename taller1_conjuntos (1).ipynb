```python
from supervenn import supervenn


A = set(range(5, 21))
B = set(x for x in range(25) if x % 2 == 0)
C = {1, 4, 8, 10, 12, 15, 18, 20}
D = set(x for x in range(2, 46) if all(x % i != 0 for i in range(2, int(x**0.5)+1)))


sets = [A, B, C, D]
labels = ['A', 'B', 'C', 'D']

supervenn(sets, labels)
```

```python
import matplotlib.pyplot as plt
from supervenn import supervenn

sym_diff_CD = C.symmetric_difference(D)
result_1 = B.intersection(sym_diff_CD)
print("Resultado 1: ", result_1)
supervenn([B, sym_diff_CD, result_1], set_annotations=['B', 'C ⊕ D', 'B ∩ (C ⊕ D)'])
plt.title("B ∩ (C ⊕ D)")

intersection_AC = A.intersection(C)
result_2 = intersection_AC.union(B)
print("Resultado 2: ", result_2)
supervenn([intersection_AC, B, result_2], set_annotations=['A ∩ C', 'B', '(A ∩ C) ∪ B'])
plt.title("(A ∩ C) ∪ B")

union_BD = B.union(D)
result_3 = union_BD.difference(C)
print("Resultado 3: ", result_3)
supervenn([union_BD, C, result_3], set_annotations=['B ∪ D', 'C', '(B ∪ D) - C'])
plt.title("(B ∪ D) - C")
plt.show()

difference_AB = A.difference(B)
intersection_AD = A.intersection(D)
result_4 = difference_AB.symmetric_difference(intersection_AD)
print("Resultado 4: ", result_4)
supervenn([difference_AB, intersection_AD, result_4], set_annotations=['A - B', 'A ∩ D', '(A - B) ⊕ (A ∩ D)'])
plt.title("(A - B) ⊕ (A ∩ D)")
plt.show()
```
